```python
import matplotlib.pyplot as plt
import numpy as np

최고기온1 = [25, 27, 30, 27, 32]
최고기온2 = [22, 23, 35, 32, 27]

plt.plot(
    최고기온1, label="year 2000", marker="o", markersize=10, linestyle="--", linewidth=3
)
plt.plot(
    최고기온2, label="year 2010", marker="o", markersize=10, linestyle="--", linewidth=3
)
plt.legend(loc="upper left")

plt.show()
```

```python
import matplotlib.pyplot as pit
import numpy as np

year = [1960, 1970, 1980, 1990, 2000, 2010, 2018]
birth = [1099294, 1006645, 862835, 649738, 634501, 470171, 326900]
pos = np.arange(len(year))
plt.xticks(pos, year)
pit.bar(pos, birth, color="g", alpha=0.5)
pit.plot(pos, birth, color="g", marker="o", linestyle="--")
pit.show()
```
