###(1)데이터 읽어오기

- **파일경로명이 다를 경우 반드시 수정**해주세요.


```python
# 시군구별 교통사고 통계.csv 화일 읽고 상위 5개 데이터 확인하기
# 변수명 = pd.read_csv(‘파일경로명’, encoding=‘인코딩방식’)
# 인코딩방식 :EUC-KR(한글이 포함된 일반적인 경우)/cp949(MS office에서 저장한 파일 형식)

import pandas as pd

accData = pd.read_csv("시군구별 교통사고 통계.csv", encoding="EUC_KR")
accData
```

###(2)데이터 정보 확인하기


```python
# 행열개수, 데이터 타입, 열의 개수 등 데이터 정보 확인하기
accData.info()
```

```python
# 숫자 데이터의 통계치 확인하기
accData.describe()
```

##2.데이터 전처리하기


###(1)열 정리하기


```python
# 열이름 변경하고 현재 테이블에 결과 반영하기
# 광역지방자치->시도
# 기초지방자치->시군구

# 변수명.rename(columns = {'열이름':'새로운 열이름'}, inplace= True)
# inplace=True 사용시 이전 열이름 더이상 사용할 수 없음

accData.rename(columns={"광역지방자치": "시도", "기초지방자치": "시군구"}, inplace=True)
accData.head()
```

```python
# [사고건수] 열을 삭제하고 원본 테이터에 반영하기

# 변수명.drop(columns=['열이름'], axis=1, inplace= True)
# inplace= True 사용 후 '사고건수'열을 더 이상 사용할 수 없음

accData.drop(columns=["사고건수"], axis=1, inplace=True)
accData
```

###(2)행 정리하기


```python
# 0번째~15번째 행 삭제하고 원본데이터에 반영 후 상위 5개 확인하기

# 변수명.drop(accData.index[], axis=0, inplace=True)

accData.drop(accData.index[0:16], axis=0, inplace=True)
accData.head()
```

```python
# 기존 index 버리고 새로운 인덱스 설정하여 원본 데이터에 반영하기

# 변수명.reset_index(drop=True, inplace=True)
# drop=True : 기존 index 열은 삭제됨

accData.reset_index(drop=True, inplace=True)
accData.head()
```

##3.데이터 재설정하기


###(1)열방향으로 테이블 결합하기


####1)테이블 결합을 위한 시군구별 부상신고자수.xlsx 화일 가져오기


```python
# 시군구별 부상신고자수.xlsx : 시도, 시군구에 따른 부상신고자수  데이터

# 변수명 = pd.read_excel('파일경로명')

부상자수 = pd.read_excel("시군구별 부상신고자수.xlsx")
부상자수.head()
```

####2)열방향으로 테이블 결합(시군구별 교통사고 통계 + 시군구별 부상신고자수)하기

---


```python
# 대전, 대구, 인천 지역은 모두 중구,서구, 동구, 남구가 있으므로 시도, 시군구를 기준으로 결합

# 변수명 = pd.merge(변수명1, 변수명2, on='결합기준열이름')
# on=['기준열1','기준열2']

mergeData = pd.merge(accData, 부상자수, on=["시도", "시군구"])
mergeData
```

###(2)행방향으로 테이블 결합하기


####1)테이블 결합을 위한 서울\_부산 교통사고 통계.xlsx 화일 가져오기


```python
# 서울_부산 교통사고 통계.xlsx : 서울과 부산지역의 교통사고 통계자료

서울부산 = pd.read_excel("서울_부산 교통사고 통계.xlsx")
서울부산.head()
```

####2)행방향으로 테이블 결합(부상신고자수가 포함된 테이블 + 서울\_부산 교통사고 통계)하기


```python
# 변수명 = pd.concat([변수명1, 변수명2])

resultData = pd.concat([mergeData, 서울부산])
resultData
```

####3)부상자수 확인하기


```python
# 부상자수 = 중상자수+경상자수+부상신고자수

# 변수명['열이름'] = 변수명['열이름'] + 변수명['열이름']

resultData["부상자수"] = (
    resultData["중상자수"] + resultData["경상자수"] + resultData["부상신고자수"]
)
resultData.head()
```

##4.데이터 분석하기(질문에 답해보기)


###(1)사망자 수가 30명이상이고 중상자가 800명 이상인 지자체는 어디인가?


```python
# 사망자수가 30명 이상인 지자체 확인하기

# 변수명[변수명['열이름']검색조건]
# 검색조건: 비교연산자를 사용
resultData[resultData["사망자수"] >= 30]
```

```python
# 사망자 수가 30명이상이고 중상자가 800명 이상인 지자체 검색하기

# 변수명[(변수명['열이름1']검색조건) 논리연산자 (변수명['열이름2']검색조건)]
# 검색조건: 비교연산자와 논리연산자를  사용
resultData[(resultData["사망자수"] >= 30) & (resultData["중상자수"] >= 800)]
```

```python
# 사망자 수가 30명이상이고 중상자가 800명 이상인 지자체 검색하기 (query)

# 변수명.query('검색 조건')
# 검색조건: 비교연산자와 논리연산자를  사용
resultData.query("사망자수 >= 30 & 중상자수 >= 800")[["사망자수", "중상자수"]]
```

###(2)사망자 수가 50명이상이거나 부상자수가 3000명 이상인 지자체는 어디인가?


```python
# 사망자 수가 50명이상이거나 부상자가 3000명 이상인 지자체 이름 검색하기 (query)

# 변수명.query('검색 조건')
# 검색조건: 비교연산자와 논리연산자를  사용
resultData.query("사망자수>=50 | 부상자수 >= 3000")[["사망자수", "부상자수"]]
```

###(3)서울과 대구에서 경상자수가 가장 많은 시군구는 어디인가?


```python
# '서울'지역의 데이터 추출하기
resultData[resultData["시도"] == "서울"]
```

```python
# query()를 사용하여'서울'지역의 데이터 추출하기
# 특정 문자열을 검색할 경우 따옴표 주의
resultData.query("시도 == '서울'")
```

```python
# '서울과 대구'지역의 데이터 추출하기
resultData[(resultData["시도"] == "시도") | (resultData["시도"] == "대구")]
```

```python
# query를 사용하여 '서울과 대구'지역의 데이터 추출하고 변수에 저장하기

서울대구 = resultData.query("시도 == '서울' | 시도 == '대구'")
서울대구
```

```python
# 서울대구 지역에서 경상자가 가장 많은 상위 10개 지자체 수 확인하기

# 변수명.sort_values(['열이름'], ascending=True)
# ascending = True:오름차순, False:내림차순, 생략:오름차순
서울대구.sort_values(["경상자수"], ascending=False).head(10)
```

###(4)사망자가 가장 많은/가장 적은 10개 지차제는 어디인가?


```python
# 사망자수가 가장 많은 10개 지자체 수 확인하기
resultData.sort_values(["사망자수"], ascending=True).head(10)
```

```python
# 사망자수가 가장 적은 10개 지자체 수 확인하기
resultData.sort_values(["사망자수"], ascending=False).head(10)
```

###(5)부상자수가 가장 많은 시도는 어디인가?


```python
# 시도별 부상자수 합계 알아보기

# 변수명.groupby('그룹열이름') [['열이름1','열이름2']]. 통계함수()

시도별부상자수 = resultData.groupby("시도")[["부상자수"]].sum()
시도별부상자수
```

```python
# 부상자수가 많은 시도 확인하기
시도별부상자수.sort_values(["부상자수"], ascending=False)
```

```python
# 시도별 '사망자수','중상자수','경상자수','부상신고자수'의 합계 알아보기
시도별 = resultData.groupby("시도")[
    ["사망자수", "중상자수", "경상자수", "부상신고자수"]
].mean()
```

###(6)사망자 평균이 가장 많은 도시는 어디인가?


```python
# 시도별 사망자수, 중상자수, 경상자수, 부상신고자수, 부상자수 평균 살펴보기
시도별
```

```python
# 시도별 사망자 평균이 많은 10개 시도 알아보기
시도별.sort_values(["사망자수"], ascending=False).head(10)
```
