```python
import matplotlib.pyplot as plt
import random

x = list(range(10))
y = [random.randint(1, 100) for _ in x]

ss = [2**n for n in range(10)]

plt.scatter(x, y, c="g", s=ss)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

matplotlib.rcParams["font.family"] = "AppleGothic"
plt.rcParams["axes.unicode_minus"] = False

# x = np.arange(0, 3.14 * 2, 0.1)
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sine")
plt.plot(x, y2, label="cosine")

plt.title("sine & cosine graph")
plt.xlabel("x value")
plt.ylabel("y value")
plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np

scores = []
for i in range(5):
    val = int(input(f"{i+1}   -> "))
    scores.append(val)

print("  :", *scores)

sumR = np.sum(scores)
aveR = np.mean(scores)
varR = np.var(scores)
stdR = np.std(scores)

print(sumR, aveR, varR, stdR)
```
