```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

matplotlib.rcParams["font.family"] = "AppleGothic"

도시 = ["부산", "대구", "인천", "광주", "대전"]
출생아수 = [1446, 10661, 14947, 7956, 7414]
혼인건수 = [11081, 7287, 10881, 4901, 5419]

pos = np.arange(len(도시))
plt.xticks(pos, 도시)

plt.bar(pos, 출생아수, label="출생아수", width=0.25)
plt.bar(pos + 0.3, 혼인건수, label="혼인건수", width=0.25)
plt.legend()

plt.title("5대 광역시 혼인건수와 출생아수")
plt.xlabel("도시")
plt.ylabel("혼인(건수) 및 출생아(수)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = list(range(10))
y = [random.randint(1, 100) for _ in x]

ss = [2**n for n in range(10)]

plt.scatter(x, y, c="g", s=ss)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

matplotlib.rcParams["font.family"] = "AppleGothic"
plt.rcParams["axes.unicode_minus"] = False

# x = np.arange(0, 3.14 * 2, 0.1)
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sine")
plt.plot(x, y2, label="cosine")

plt.title("sine & cosine graph")
plt.xlabel("x value")
plt.ylabel("y value")
plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np

scores = []
for i in range(5):
    val = int(input(f"{i+1}번 성적 입력 -> "))
    scores.append(val)

print("입력된 전체 점수:", *scores)

sumR = np.sum(scores)
aveR = np.mean(scores)
varR = np.var(scores)
stdR = np.std(scores)

print(sumR, aveR, varR, stdR)
```
