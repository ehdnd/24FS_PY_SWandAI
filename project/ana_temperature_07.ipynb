```python
start_date = pd.Timestamp("2024-07-01")
date_idx = start_date + pd.to_timedelta(temp_1hour_07["day"] - 1, unit="d")
date_idx
```

```python
time_idx = pd.to_timedelta((temp_1hour_07["hour"] // 100).astype(int), unit="h")
time_idx
```

```python
temp_1hour_07.head()
```

```python
temp_1hour_07.set_index([date_idx, time_idx], inplace=True)
temp = temp_1hour_07[["temp"]]
temp
```

```python
avg_temp = temp.groupby(level=0).mean()
avg_temp.rename(columns={"temp": "avg_temp"}, inplace=True)
avg_temp
```

```python
max_temp = temp.groupby(level=0).max()
max_temp.rename(columns={"temp": "max_temp"}, inplace=True)
```

```python
min_temp = temp.groupby(level=0).min()
min_temp.rename(columns={"temp": "min_temp"}, inplace=True)
min_temp
```

```python
total_temp = pd.concat([avg_temp, max_temp, min_temp], axis=1)
total_temp
```

```python
total_temp.to_csv("07_avg_max_min_temp.csv")
```
