```python
import pandas as pd
import numpy as np
```

```python
historical = pd.read_csv(
    "./data/historical_data/historical.csv",
    encoding="cp949",
    header=0,
    usecols=[2, 3, 4, 6, 8],
    names=["day", "avg_temp", "min_temp", "max_temp", "rain"],
)
historical
```

```python
historical["rain"] = historical["rain"].astype(bool)
historical.loc[historical["rain"] > 0, "rain"] = True
historical.loc[historical["rain"] <= 0, "rain"] = False
historical["rain"] = historical["rain"].fillna(False)
historical["day"] = pd.to_datetime(historical["day"])
historical.set_index("day", inplace=True)
historical
```

```python
historical.to_csv("historical_avg_min_max_rain.csv")
```
